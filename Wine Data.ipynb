# Part 2. Text Analysis

## 2.1 extract high requency words in description

```python
!pip install nltk
```

```python
import re
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.probability import *
from itertools import chain
#from tqdm import tqdm
import codecs
```

```python
with open('stopwords.txt') as f:
    stop_words = f.read().splitlines()
stop_words = set(stop_words)
stop_words
```

```python
tokenizer = RegexpTokenizer(r"\w+(?:[-']\w+)?")
sentencelist=[]
for words in w["description"]:
    sentence=[a for a in tokenizer.tokenize(words) if a not in stop_words]
    sentencelist.append(sentence)
tokenizedwords=[word for word in sentence for sentence in sentencelist]
tokenizedwords
```

```python
from itertools import takewhile
fdist = FreqDist(tokenizedwords)
common_words=fdist.most_common()
high_freq_comm=dict(takewhile(lambda x:x[1]>5000,common_words))
key_word=high_freq_comm.keys()
sorted_high_common={}
for key in sorted(key_word):
    sorted_high_common[key]=high_freq_comm[key]
sorted_high_common
```

```python
with open('HighFreq.txt','a')as file:
    file.write(str(sorted_high_common))
```

## 2.2 Find key words for describing Shiraz using TF-IDF

```python
shirazsentences=w[w['variety']=='Shiraz']['description']
shirazsentences
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf = TfidfVectorizer(analyzer='word', stop_words = 'english')
x=tf_idf.fit_transform(shirazsentences)
idf=tf_idf.idf_
tfids_values=dict(zip(tf_idf.get_feature_names(),idf))
tfids_values
```

```python
score=list((key,value.round(2)) for (key,value) in tfids_values.items() if value>0.4)
score.sort()
score
```

```python
with open("ShirazKey.txt",'w') as file:
    file.write(str(score))
```
